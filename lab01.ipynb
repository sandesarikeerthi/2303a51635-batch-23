```python
from graphviz import Digraph

def factorial_with_viz(n, dot=None):
    if dot is None:
        dot = Digraph()
        dot.attr('node', shape='circle')

    if n == 0:
        dot.node(str(id(n)), '1')
        return 1, dot
    else:
        parent_id = str(id(n))
        dot.node(parent_id, f'f({n})')
        result, dot = factorial_with_viz(n - 1, dot)
        child_id = str(id(n-1))
        dot.edge(parent_id, child_id)
        return n * result, dot

x = int(input('Enter a number (<=10): '))
if x <= 10:
    _, dot = factorial_with_viz(x)
    display(dot)
    result, _ = factorial_with_viz(x)
    print(f'Factorial of {x} is {result}')
else:
    print("Input exceeded")
```
